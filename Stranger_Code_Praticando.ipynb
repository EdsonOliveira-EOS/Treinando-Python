## ðŸŸ¥ **NÃ­vel 1: SeguranÃ§a do LaboratÃ³rio** (Encapsulamento e LÃ³gica)

O laboratÃ³rio de Hawkins foi reativado em segredo pelo exÃ©rcito. VocÃª encontrou um cÃ³digo anotado no bolso de um cientista, mas o sistema bloqueia qualquer tentativa errada.

**Desafio:**
Crie uma classe `PainelSeguranca`.
1.  O `__init__` define uma `senha_correta` e o status inicial como **"Bloqueado"**.
2.  O mÃ©todo `tentar_acesso(senha)` compara a entrada:
    * Se acertar: Muda status para **"Liberado"** e avisa que o acesso foi concedido.
    * Se errar: MantÃ©m **"Bloqueado"** e aciona um alerta de seguranÃ§a.
3.  Um mÃ©todo `verificar_status()` para checar se a porta abriu.


```python
class PainelSeguranÃ§a():
  def __init__ (self, senhacorreta, estado = 'Bloqueado'):
    self.senhacorreta = senhacorreta
    self.estado = estado

  def tentar_acesso(self, senha):
    if senha == self.senhacorreta:
      self.estado = 'Liberado'
      print ('Click. Senha Correta')

    else:
      self.estado = 'Bloqueado'
      print ('Click. Senha Incorreta')

  def verificar_status(self):
    print(f'O estado atual Ã© {self.estado}')

porta = PainelSeguranÃ§a('NINA_PROJECT')
porta.tentar_acesso("1234") # bloqueado
porta.verificar_status()
porta.tentar_acesso("NINA_PROJECT") # liberado
porta.verificar_status()
```

## ðŸŸ§ **NÃ­vel 2: O Transe da Max (Listas e Estado)**

**Desafio:**
Crie uma classe `Walkman`.

2.  O mÃ©todo `tocar_playlist_inteira()`:

  * Cada mÃºsica gasta 1 de bateria.
  * Se a bateria acabar (0) durante a playlist, o som para e retorna **SilÃªncio... A bateria morreu!**.
  * Se tocar tudo, retorna **Playlist finalizada com sucesso!**.

```python
class Walkman():
  def __init__ (self, fitas, bateria = 5):
    self.listafitas = fitas
    self.bateria = bateria

  def tocar_playlist_inteira(self):
    for musica in self.listafitas:
      self.bateria -= 1

    if self.bateria >= 0:
      print ('Playlist finalizada com sucesso!')

    else:
      print ('SilÃªncio... A bateria morreu!.')

fitas = ["Running Up That Hill", "Psycho Killer", "Should I Stay", "Material Girl"]
sony = Walkman(fitas)
sony.tocar_playlist_inteira() # Playlist finalizada com sucesso!.

fitas2 = ['LolÃ³', 'Elissan', 'Somewhere only we know', 'Lanadeyray', 'Labubu', 'Labirosca']
sony2 = Walkman(fitas2)
sony2.tocar_playlist_inteira()
```

## ðŸŸ¨ **NÃ­vel 3: O Arsenal do Steve** (InteraÃ§Ã£o entre Objetos)

As criaturas estÃ£o vindo e o taco de beisebol comum nÃ£o vai causar nem arranhÃµes na pele dura dos monstros. Steve Harrington achou uma caixa de pregos enferrujados na garagem. Ã‰ hora de fazer um *upgrade* manual antes que a horda chegue.

**A MissÃ£o:**
Crie um sistema de Crafting onde um objeto modifica o outro.
1.  Crie a classe `CaixaDePregos` com atributo `quantidade`.
2.  Crie a classe `Bastao` com atributo `dano` (10) e `nome` ("Taco").
3.  No `Bastao`, crie o mÃ©todo `aplicar_upgrade(caixa)`:
    * O mÃ©todo deve receber o **objeto** `caixa` como parÃ¢metro.
    * Se `caixa.quantidade > 0`: Aumente o dano do bastÃ£o (+15), mude o nome para "Taco com Pregos" e reduza 1 da quantidade da caixa.
    * Se nÃ£o tiver pregos: Retorne *"Sem recursos disponÃ­veis."*.



```python
class Bastao():
  def __init__ (self):
    self.dano = 10
    self.nome = 'Taco'

  def aplicar_upgrade(self, caixa):
    if caixa.quantidade > 0:
      self.nome = 'Taco com pregos'
      self.dano += 15
      caixa.quantidade -= 1

    else:
      print ('Sem recursos disponÃ­veis.')

class CaixaDePregos():
  def __init__ (self, quantidade):
    self.quantidade = quantidade


taco = Bastao()
caixa = CaixaDePregos(quantidade=10)

print(f"Dano Atual: {taco.dano}")
taco.aplicar_upgrade(caixa)
print(f"Novo Dano: {taco.dano}")

print(f'Quantidade Pregos: {caixa.quantidade}')
```

## Utilizando a classe (input)

```python
taco = Bastao()
caixa = CaixaDePregos(quantidade=10)

print(f"Dano Atual: {taco.dano}")
taco.aplicar_upgrade(caixa)
print(f"Novo Dano: {taco.dano}")
```

Dustin percebeu que o Mundo Invertido estÃ¡ evoluindo. O que antes eram apenas animais selvagens (Demodogs) agora sÃ£o generais conscientes (Vecna). Seu cÃ³digo precisa diferenciar o comportamento instintivo de uma besta do ataque psÃ­quico calculado do vilÃ£o, reaproveitando a estrutura bÃ¡sica de um monstro.

**A MissÃ£o:**
Use HeranÃ§a para criar inimigos diferentes.
1.  Classe MÃ£e `Monstro`: Tem atributo `nome` e mÃ©todo `atacar()` que retorna *"Ataque genÃ©rico"*.
2.  Classe Filha `Demodog`: Herda de Monstro.
    * Sobrescreva o mÃ©todo `atacar()`. Use `super().atacar()` para pegar a frase original e concatene com *"Mordida violenta!"*.
3.  Classe Filha `Vecna`: Herda de Monstro.
    * Sobrescreva **totalmente** o mÃ©todo `atacar()` (sem super) para retornar *"Invadindo sua mente... Dano PsÃ­quico!"*.


```python
class Monstro ():
  def __init__ (self, nome):
    self.nome = nome

  def atacar(self):
    return f'{self.nome} usou o ataque genÃ©rico: '

class Demodog(Monstro):
  def atacar(self):
    return super().atacar() + 'Mordida Violenta'

class Vecna(Monstro):
  def atacar(self):
    return f'{self.nome} estÃ¡ invadindo sua mente... Dano PsÃ­quico!'

soldado = Demodog("Dart")
general = Vecna("001")

print(soldado.atacar())
print(general.atacar())
```

## â¬› **NÃ­vel 5: O Ataque Sincronizado** (Lista de Objetos)

O portal central estÃ¡ aberto no meio de Hawkins. Hopper, Eleven, Mike e Nancy estÃ£o posicionados, mas Vecna Ã© forte demais para ser derrotado por um sÃ³. O grupo precisa atacar em perfeita sincronia: se a soma dos poderes deles nÃ£o for suficiente para superar a defesa do portal, Hawkins cairÃ¡.

**A MissÃ£o:**
Simule a batalha final usando Listas e Objetos.
1.  Crie a classe `Heroi` com atributos `nome` e `poder` (int).
2.  Crie a classe `Portal` com atributo `defesa` (int).
3.  No `Portal`, crie o mÃ©todo `receber_ataque_coordenado(lista_de_herois)`:
    * O mÃ©todo recebe uma **lista** contendo vÃ¡rios objetos `Heroi`.
    * Some o `poder` de todos os herÃ³is da lista.
    * Compare: Se `soma_poderes >= self.defesa`, retorne *"Sucesso! O portal foi fechado."*. Caso contrÃ¡rio, retorne *"Falha... O grupo pereceu."*.



```python
class Heroi():
  def __init__ (self, nome, poder):
    self.nome = nome
    self.poder = poder

class Portal():
  def __init__ (self, defesa):
    self.defesa = defesa

  def receber_ataque_coordenado(self, lista_de_herois):
    poder_total = sum(len(heroi.nome) for heroi in lista_de_herois)

    for heroi in lista_de_herois:
      heroi.poder += poder_total

    if sum(heroi.poder for heroi in lista_de_herois) >= self.defesa:
      return 'Sucesso! O portal foi fechado'

    else:
      return 'Falha.. o grupo pereceu'

el = Heroi("Eleven", 100)
hopper = Heroi("Hopper", 60)
grupo_ataque = [el, hopper]

portal_central = Portal(defesa=150)
print(portal_central.receber_ataque_coordenado(grupo_ataque))
```

```python
el = Heroi("Eleven", 100)
hopper = Heroi("Hopper", 60)
grupo_ataque = [el, hopper]

portal_central = Portal(defesa=150)
print(portal_central.receber_ataque_coordenado(grupo_ataque))
```
